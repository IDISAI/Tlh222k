# Chạy Code Cell & Run All

Kernel là tiến trình Python thật chạy trong Docker.

## Chạy từng cell

- Bấm **▷** (tròn, bên trái cell) → chạy cell đó
- `Shift+Enter` → chạy và chuyển xuống cell tiếp theo
- Output stream **real-time** (từng dòng print hiện ngay)

## Run All

- Bấm **Run All** → chạy toàn bộ code cell từ trên xuống
- Số **In [N]** cho biết thứ tự thực thi

```python
import time

print('Bắt đầu...')
for i in range(1, 6):
    time.sleep(0.3)
    print(f'Bước {i}/5 — streaming real-time!')
print('Xong!')
```

## Kernel status

| Màu | Nghĩa |
|---|---|
| 🟢 xanh | idle — sẵn sàng |
| 🟡 nhấp nháy | starting |
| 🟡 tĩnh | busy — đang chạy |
| 🔴 đỏ | error |
