# Trực quan hóa với Matplotlib

Nhìn thấy dữ liệu trước khi mô hình hóa. Biểu đồ hiện ngay dưới ô code.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.legend()
plt.title("Đường sin và cos")
plt.show()
```

## Scatter — quan hệ giữa 2 biến

```python
rng = np.random.default_rng(42)
gio_hoc = rng.uniform(0, 10, 50)
diem = 5 + 0.45 * gio_hoc + rng.normal(0, 0.5, 50)

plt.scatter(gio_hoc, diem, alpha=0.7)
plt.xlabel("Giờ tự học mỗi tuần")
plt.ylabel("Điểm")
plt.title("Học nhiều hơn, điểm cao hơn?")
plt.show()
```

## Histogram — phân phối của 1 biến

```python
data = rng.normal(loc=170, scale=6, size=1000)   # chiều cao (cm)
plt.hist(data, bins=30, edgecolor="black")
plt.xlabel("Chiều cao (cm)")
plt.title("Phân phối chuẩn quanh 170cm")
plt.show()
```
