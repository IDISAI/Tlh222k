# Mạng nơ-ron từ zero

Đích của lộ trình: tự cài mạng nơ-ron 2 lớp bằng NumPy học hàm **XOR** — bài toán mà một đường thẳng không thể giải.

```python
import numpy as np

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=float)
y = np.array([[0], [1], [1], [0]], dtype=float)   # XOR
print("XOR: đầu vào khác nhau -> 1, giống nhau -> 0")
```

## Kiến trúc: 2 → 4 → 1

Lớp ẩn 4 neuron với sigmoid cho mô hình khả năng "bẻ cong" ranh giới quyết định.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

rng = np.random.default_rng(1)
W1 = rng.normal(0, 1, (2, 4))
b1 = np.zeros(4)
W2 = rng.normal(0, 1, (4, 1))
b2 = np.zeros(1)
```

## Train với backpropagation

Lan truyền xuôi tính dự đoán; lan truyền ngược tính gradient qua quy tắc chuỗi.

```python
lr = 1.0
for buoc in range(3000):
    # lan truyền xuôi
    h = sigmoid(X @ W1 + b1)
    y_hat = sigmoid(h @ W2 + b2)

    # lan truyền ngược (đạo hàm của binary cross-entropy + sigmoid)
    d_out = y_hat - y
    dW2 = h.T @ d_out / len(X)
    db2 = d_out.mean(axis=0)
    d_hidden = (d_out @ W2.T) * h * (1 - h)
    dW1 = X.T @ d_hidden / len(X)
    db1 = d_hidden.mean(axis=0)

    W1 -= lr * dW1; b1 -= lr * db1
    W2 -= lr * dW2; b2 -= lr * db2

    if buoc % 600 == 0:
        loss = -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat)).mean()
        print(f"bước {buoc:4d}: loss = {loss:.4f}")
```

## Kiểm tra kết quả

```python
y_hat = sigmoid(sigmoid(X @ W1 + b1) @ W2 + b2)
for dau_vao, du_doan, dich in zip(X, y_hat, y):
    print(f"{dau_vao} -> {du_doan[0]:.3f} (đích: {int(dich[0])})")
```
