# Pandas cơ bản

Pandas = bảng tính lập trình được. `DataFrame` là cấu trúc trung tâm của mọi dự án dữ liệu.

```python
import pandas as pd

df = pd.DataFrame({
    "ten": ["Lan", "Minh", "Huy", "Trang", "Quan"],
    "nganh": ["AI", "AI", "Web", "Data", "Web"],
    "diem": [8.5, 7.0, 6.5, 9.0, 7.5],
})
df
```

## Lọc dữ liệu

Điều kiện boolean trả về đúng các dòng cần tìm.

```python
print(df[df["diem"] >= 7.5])
print(df[(df["nganh"] == "AI") & (df["diem"] > 7)])
```

## Cột mới và thống kê mô tả

```python
df["xep_loai"] = df["diem"].apply(lambda d: "Giỏi" if d >= 8 else "Khá")
df.describe()
```

## `groupby` — chia nhóm rồi tổng hợp

```python
df.groupby("nganh")["diem"].agg(["mean", "max", "count"])
```
