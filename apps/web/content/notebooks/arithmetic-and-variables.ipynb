## Printing

One of the simplest (and most important!) tasks you can ask a computer to do is to print a message. We do this with the `print()` function:

```python
print("Hello, world!")
```

## Arithmetic

Python is great at arithmetic. The table below shows the most common operators:

- `+` addition
- `-` subtraction
- `*` multiplication
- `/` division
- `**` exponent

For example, to add two numbers:

```python
print(1 + 2)
print(9 - 5)
print(2 ** 5)
```

## Comments

We use **comments** to annotate what code is doing. They start with `#`, and Python ignores them when running the code:

```python
# Multiply 3 by 2
print(3 * 2)
```

## Variables

So far, you have used code to make a calculation and print the result, but the result isn't saved anywhere. **Variables** let you save a value to work with later.

### Creating variables

To create a variable, use `=` with the variable name on the left and the value on the right:

```python
test_var = 4 + 5
print(test_var)

# Create a variable we'll keep using below
hours_per_day = 24
```

### Manipulating variables

Once a variable exists you can use it in new calculations. For instance, the number of seconds in a year:

```python
# Calculate the number of seconds in a year
total_secs = 60 * 60 * hours_per_day * 365
total_secs
```

### Using multiple variables

Real analyses combine many variables. Here we use `pandas` to organize a few of them into a table:

```python
import pandas as pd

durations = pd.DataFrame({
    "unit": ["day", "week", "year"],
    "hours": [hours_per_day, hours_per_day * 7, hours_per_day * 365],
})
durations
```

## Debugging

One common error when working with variables is to accidentally misspell a name. Say we mistype `hours_per_day` as `hours_per_dy`:

```python
print(hours_per_dy)
```

Python raises a `NameError` because `hours_per_dy` was never defined. Fixing the spelling makes the code run:

```python
print(hours_per_day)
```
