# Biểu đồ Matplotlib

Output hình ảnh PNG — vẽ biểu đồ và thấy ngay trong cell.

## Cách hoạt động

1. Cell chạy matplotlib
2. Kernel render figure → **PNG base64**
3. Proxy gửi về browser qua WebSocket (giới hạn 64 MB)
4. Browser hiển thị ảnh trong cell output

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, np.sin(x), color='#6366f1', linewidth=2, label='sin(x)')
axes[0].plot(x, np.cos(x), color='#f59e0b', linewidth=2, label='cos(x)')
axes[0].set_title('Hàm lượng giác')
axes[0].legend()
axes[0].grid(alpha=0.3)

axes[1].bar(['A', 'B', 'C', 'D'], [23, 45, 12, 67],
            color=['#6366f1','#8b5cf6','#a78bfa','#c4b5fd'])
axes[1].set_title('Bar Chart')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
n = 300
x = np.random.randn(n)
y = 2 * x + np.random.randn(n) * 0.8

plt.figure(figsize=(7, 5))
plt.scatter(x, y, c=np.abs(x + y), cmap='viridis', alpha=0.7, s=40)
plt.colorbar(label='|x + y|')
plt.title('Scatter plot')
plt.tight_layout()
plt.show()
```
