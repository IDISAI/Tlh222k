## 1. Biến và kiểu dữ liệu

Python không cần khai báo kiểu — gán là xong.

```python
ten = "Lan"          # chuỗi (str)
tuoi = 21             # số nguyên (int)
diem = 8.5            # số thực (float)
dang_hoc = True       # boolean

print(ten, tuoi, diem, dang_hoc)
print(type(ten), type(tuoi))
```

## 2. Danh sách và dictionary

Hai cấu trúc dữ liệu bạn sẽ dùng mỗi ngày.

```python
diem_thi = [7.5, 8.0, 9.25, 6.75]
print("Cao nhất:", max(diem_thi))
print("Trung bình:", sum(diem_thi) / len(diem_thi))

sinh_vien = {"ten": "Lan", "nganh": "AI", "nam": 2}
print(sinh_vien["nganh"])
```

## 3. Điều khiển luồng: `if` và `for`

```python
for d in diem_thi:
    if d >= 8:
        xep_loai = "Giỏi"
    elif d >= 6.5:
        xep_loai = "Khá"
    else:
        xep_loai = "Trung bình"
    print(f"{d} -> {xep_loai}")
```

## 4. Hàm

Gói logic vào hàm để tái sử dụng.

```python
def xep_loai(diem):
    """Xếp loại theo thang điểm 10."""
    if diem >= 8:
        return "Giỏi"
    if diem >= 6.5:
        return "Khá"
    return "Trung bình"

print([xep_loai(d) for d in diem_thi])
```
