# Hồi quy tuyến tính từ zero

Bài ML đầu tiên: dự đoán `y` từ `x` bằng đường thẳng `y = w*x + b`, tự cài **gradient descent** — không dùng thư viện ML nào.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)
x = rng.uniform(0, 10, 80)
y_that = 2.5 * x + 7          # quy luật thật (mô hình phải tự tìm ra)
y = y_that + rng.normal(0, 2, 80)

plt.scatter(x, y, alpha=0.6)
plt.title("Dữ liệu: y ≈ 2.5x + 7 + nhiễu")
plt.show()
```

## Gradient descent

Lặp: dự đoán → đo lỗi (MSE) → chỉnh `w, b` ngược hướng gradient.

```python
w, b = 0.0, 0.0
lr = 0.01

for buoc in range(1000):
    y_du_doan = w * x + b
    loi = y_du_doan - y
    # gradient của MSE theo w và b
    dw = 2 * (loi * x).mean()
    db = 2 * loi.mean()
    w -= lr * dw
    b -= lr * db
    if buoc % 200 == 0:
        print(f"bước {buoc:3d}: MSE = {(loi ** 2).mean():.3f}, w = {w:.2f}, b = {b:.2f}")

print(f"\nKết quả: y = {w:.2f}x + {b:.2f}  (quy luật thật: y = 2.5x + 7)")
```

## Nhìn mô hình đã học

```python
plt.scatter(x, y, alpha=0.5, label="dữ liệu")
xs = np.linspace(0, 10, 100)
plt.plot(xs, w * xs + b, color="red", linewidth=2, label=f"mô hình: {w:.2f}x + {b:.2f}")
plt.legend()
plt.show()
```
