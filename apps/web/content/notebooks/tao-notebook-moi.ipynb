# Tạo Notebook Mới

Hướng dẫn tạo và quản lý notebook trong Admin Editor.

## Cách tạo

1. Vào **Admin** → `/notebooks`
2. Bấm **+ New Notebook**
3. Nhập tên → bấm **Tạo**
4. Trang chuyển sang editor với notebook mới.

Notebook được lưu ngay với tiêu đề bạn nhập.
