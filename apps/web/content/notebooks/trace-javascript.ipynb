# Visualize JavaScript execution

Run a cell, then choose **Visualize execution** to step through it one line at a time and watch variables, the call stack, the heap, and output change.

Running these cells needs the kernel server (`NEXT_PUBLIC_KERNEL_SERVER_URL`); the visualizer itself runs entirely in your browser.

## Variables and output

Every step of the visualizer shows the value of each variable at that moment.

```javascript
const price = 12
const quantity = 3
const total = price * quantity
console.log("total", total)
```

## Objects and arrays live on the heap

Two names can point at the same object. Step through this cell and watch both `cart` and `alias` reference the same heap node.

```javascript
const cart = { items: [], total: 0 }
const alias = cart
alias.items.push("apple")
alias.items.push("pear")
cart.total = cart.items.length
console.log(cart.items.join(", "), cart.total)
```

## Functions and the call stack

Calling a function pushes a frame. The stack panel shows each frame and its own variables.

```javascript
function double(value) {
  const doubled = value * 2
  return doubled
}

function sumDoubled(numbers) {
  let total = 0
  for (const number of numbers) {
    total = total + double(number)
  }
  return total
}

const answer = sumDoubled([1, 2, 3])
console.log("answer", answer)
```
