# Tạo Exercise Notebook

Exercise là notebook companion — hiển thị ở **Tab Exercise** trên `/learn`.

## Convention slug

```
Tutorial slug:  python-co-ban
Exercise slug:  python-co-ban-exercise
```

Web tự load `{slug}-exercise` từ kernel-server.

## Quy trình

1. Admin → `/notebooks` → Tạo notebook `{tutorial-slug}-exercise`
2. Soạn bài tập:
   - Markdown giải thích yêu cầu
   - Code cell `# YOUR CODE HERE`
   - Code cell kiểm tra
3. **Xuất bản** → Tab Exercise tự kích hoạt!

```
┌─────────┐┌──────────┐
│ Tutorial ││ Exercise │  ← kích hoạt khi có {slug}-exercise
└─────────┘└━━━━━━━━━━┘
│ ## Bài tập 1: ...   |
│ [ ]: # YOUR CODE    |
└─────────────────────┘
```
