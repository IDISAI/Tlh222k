# Xuất Bản Notebook

Xuất bản = cho người dùng web truy cập notebook qua `/learn/[slug]`.

## Quy trình

1. Chỉnh sửa notebook trong Admin
2. Bấm **Xuất bản** (góc phải toolbar)
3. Popover hiện:
   - Xác nhận notebook đang public
   - URL `/learn/[slug]` — click để select, bấm 📋 copy
   - Nút **Hủy xuất bản** để gỡ

```
┌───────────────────────────────────┐
│ 🌐 Notebook đang hiển thị công khai│
│ ┌─────────────────────────┐  📋   │
│ │ localhost:3000/learn/.. │       │
│ └─────────────────────────┘       │
│ [       Hủy xuất bản       ]      │
└───────────────────────────────────┘
```

## Lưu ý

- Slug không thay đổi được sau tạo
- Published notebook: API không cần auth
- Draft: chỉ admin xem
- Thay đổi sau publish → autosave → **cập nhật ngay** trên /learn
