# Interrupt & Restart Kernel

## Interrupt — Dừng cell đang chạy

- Bấm **⏹ Dừng** khi kernel đang `busy`
- Tương đương `Ctrl+C` trong terminal
- Biến đã tính **vẫn còn** trong bộ nhớ

```python
import time
# Chạy cell này rồi bấm Interrupt
print('Bắt đầu vòng lặp...')
for i in range(9999):
    print(f'  {i}', end='\r')
    time.sleep(0.1)
```

## Restart — Reset toàn bộ kernel

- Bấm **⟳ Restart** → kernel mới, **xóa toàn bộ biến**
- Output đã render vẫn hiển thị
- Dùng trước "Run All" để đảm bảo kết quả nhất quán
