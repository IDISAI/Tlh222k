# Code Cell & Phím Tắt

CodeMirror 6 với Python syntax highlight, auto-indent và bracket matching.

## Phím tắt trong cell

| Phím | Hành động |
|---|---|
| `Shift+Enter` | Chạy → xuống cell tiếp |
| `Tab` | Indent |
| `Ctrl+/` | Toggle comment |
| `Ctrl+Z` | Undo text |

## Phím tắt cấu trúc (ngoài cell)

| Phím | Hành động |
|---|---|
| `Ctrl+Z` | Undo cấu trúc |
| `Ctrl+Shift+Z` | Redo cấu trúc |

⚠️ Undo text (trong cell) và undo cấu trúc là **hai hệ thống riêng**.

```python
data = [1, 2, 3, 4, 5]
tong = sum(data)
tb = tong / len(data)
print(f'Tổng: {tong}, Trung bình: {tb}')
```

## Accent bar & selection UI

- Cell selected → **thanh xanh** bên trái (Colab style)
- `ring-2` + `shadow` khi selected
- Hover → viền mờ hiện
- Action toolbar (↑↓⧉🗑T) chỉ hover hoặc code cell selected
