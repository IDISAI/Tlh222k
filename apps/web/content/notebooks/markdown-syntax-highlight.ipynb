# Markdown Cells với Syntax Highlight

Dùng **CodeMirror 6** khi edit — heading/bold/code được highlight ngay khi gõ.

## Cách dùng

1. Click markdown cell đang render → **edit mode**
2. Gõ markdown — highlight hiện ngay
3. Preview song song ở nửa phải (màn hình ≥ md)
4. `Shift+Enter` hoặc **Đóng** → render

**Toolbar định dạng**: H B *I* `<>` 🔗 📷 ❝ ≡ 1.

## Ví dụ

**Đậm**, *nghiêng*, `inline code`

```python
def chao(ten):
    return f'Xin chào, {ten}!'
```
