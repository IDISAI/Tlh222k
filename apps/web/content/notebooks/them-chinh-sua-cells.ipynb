# Thêm & Chỉnh Sửa Cells

Notebook gồm hai loại cell: **Code** và **Markdown**.

## Thêm cell

- **Toolbar**: bấm `Code` hoặc `Markdown` → thêm bên dưới cell đang chọn.
- **Divider**: hover vào khoảng trống giữa 2 cell → hiện `+ Code` / `+ Text`.
- Phím tắt: `Ctrl+Z` / `Ctrl+Shift+Z` để undo/redo cấu trúc.

## Di chuyển & xóa

Hover lên cell → action toolbar hiện phía trên cell:

| Nút | Hành động |
|---|---|
| ↑ / ↓ | Di chuyển lên/xuống |
| ⧉ | Nhân bản |
| 🗑 | Xóa |
| T | Chuyển Code ↔ Markdown |

## Undo / Redo cấu trúc

- Nút **↶ ↷** trên toolbar hoặc `Ctrl+Z` / `Ctrl+Shift+Z`
- Undo/redo **thêm/xóa/di chuyển/đổi loại** cell
- Lịch sử 100 bước; gõ liên tục trong 1 cell gộp 1 bước
