# NumPy cơ bản

NumPy là nền móng của mọi thư viện AI: mảng nhiều chiều + phép toán vector hóa nhanh gấp trăm lần vòng lặp Python.

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
print(a, a.dtype, a.shape)
```

## Vectorization

Phép toán áp dụng lên **cả mảng** một lúc — không cần vòng lặp.

```python
print(a * 2)        # nhân từng phần tử
print(a ** 2)       # bình phương
print(a + np.array([10, 20, 30, 40, 50]))
```

## Mảng 2 chiều (ma trận)

```python
m = np.arange(12).reshape(3, 4)
print(m)
print("hàng 0:", m[0])
print("cột 1:", m[:, 1])
print("tổng theo cột:", m.sum(axis=0))
```

## Broadcasting

NumPy tự "kéo giãn" mảng nhỏ cho khớp mảng lớn — quy tắc quan trọng nhất khi làm việc với dữ liệu.

```python
diem = np.array([[7, 8], [5, 9], [6, 6]])   # 3 sinh viên x 2 môn
he_so = np.array([0.4, 0.6])                 # trọng số 2 môn
diem_tb = (diem * he_so).sum(axis=1)
print(diem_tb)
```
