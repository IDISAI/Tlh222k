# Xử Lý Lỗi & Traceback

Khi code lỗi, notebook hiển thị traceback có màu ANSI với nền đỏ nhạt.

## Đọc traceback

- `----> N` chỉ đúng **dòng lỗi**
- Tên lỗi (NameError, TypeError…) ở cuối
- ANSI color: đỏ = lỗi path, xanh = frame code

```python
# NameError — biến chưa khai báo
print(bien_chua_khai_bao)
```

## Sửa lỗi

1. Đọc `----> N` → biết dòng lỗi
2. Sửa code trong cell
3. Chạy lại (`Shift+Enter`)
4. Lỗi do biến từ cell trước → **Restart + Run All**
